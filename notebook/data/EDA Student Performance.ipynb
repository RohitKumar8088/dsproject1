### 3.7 Adding columns for "Total Score" and "Average"

```python
df['total_score']=df['math_score']+df['reading_score']+df['writing_score']
df['average']=df['total_score']/3
df.head()
```

```python
math_full=df[df['math_score']==100]['average'].count()
reading_full=df[df['reading_score']==100]['average'].count()
writing_full=df[df['writing_score']==100]['average'].count()

print(f'Number of students with full marks in Maths : {math_full}')
print(f'Number of students with full marks in Reading : {reading_full}')
print(f'Number of students with full marks in Writing : {writing_full}')
```

```python
math_lessORequal_20=df[df['math_score']<=20]['average'].count()
reading_lessORequal_20=df[df['reading_score']<=20]['average'].count()
writing_lessORequal_20=df[df['writing_score']<=20]['average'].count()

print(f'Number of students with less than or equal marks in Maths : {math_lessORequal_20}')
print(f'Number of students with less than or equal marks in Reading : {reading_lessORequal_20}')
print(f'Number of students with less than or equal marks in Writing : {writing_lessORequal_20}')
```

### Insights
    -> From above values we get students have performed the worst in Maths
    -> Best performance in Reading section
